```python
import csv

def oracle_2bit_multiplier(a1, a0, b1, b0):
    """
    Oracle function: computes the expected 2-bit multiplication result.
    Logic: (A1 A0) * (B1 B0) = P3 P2 P1 P0
    """
    # Convert binary inputs to decimal
    A = (a1 << 1) | a0
    B = (b1 << 1) | b0

    # Perform mathematical multiplication
    P = A * B

    # Extract the decimal result back to 4-bit binary (p3 p2 p1 p0)
    p0 = P & 1
    p1 = (P >> 1) & 1
    p2 = (P >> 2) & 1
    p3 = (P >> 3) & 1

    return p0, p1, p2, p3

def validate_truth_table(filename):
    print(f"Starting Oracle validation for 2-bit Multiplier using file: {filename}...\n")
    print("-" * 65)

    try:
        with open(filename, mode='r', encoding='utf-8') as f:
            reader = csv.DictReader(f)

            # Clean up potential spaces in column names just in case
            reader.fieldnames = [name.strip() for name in reader.fieldnames]

            mismatch_found = False
            row_num = 0

            for row in reader:
                row_num += 1

                # Read input pins
                a0 = int(row['A0'].strip())
                a1 = int(row['A1'].strip())
                b0 = int(row['B0'].strip())
                b1 = int(row['B1'].strip())

                # Read actual outputs generated by LLM's circuit
                actual_p0 = int(row['P0'].strip())
                actual_p1 = int(row['P1'].strip())
                actual_p2 = int(row['P2'].strip())
                actual_p3 = int(row['P3'].strip())

                # Get expected outputs from Oracle
                exp_p0, exp_p1, exp_p2, exp_p3 = oracle_2bit_multiplier(a1, a0, b1, b0)

                # Compare actual vs expected
                if (actual_p0 != exp_p0 or actual_p1 != exp_p1 or
                    actual_p2 != exp_p2 or actual_p3 != exp_p3):

                    print(f"Mismatch found!")
                    print(f"First failing input vector occurred at row {row_num}:")
                    print(f"Input vector: A(A1 A0)={a1}{a0}, B(B1 B0)={b1}{b0} (Decimal: {a1<<1 | a0} * {b1<<1 | b0})")
                    print(f"Expected output (P3 P2 P1 P0): {exp_p3} {exp_p2} {exp_p1} {exp_p0}")
                    print(f"Actual output   (P3 P2 P1 P0): {actual_p3} {actual_p2} {actual_p1} {actual_p0}")
                    print("-" * 65)

                    # Explain likely cause as required by Part II
                    print("Likely cause analysis for report:")
                    print("CNF issue: The LLM likely failed to synthesize the proper carry-over logic or intermediate AND/ADD operations within the multiplier.")

                    mismatch_found = True
                    break # Stop at the first failure

            if not mismatch_found:
                print(f"Validation passed! All {row_num} rows match the expected oracle outputs perfectly.")
                print("Conclusion for Part II & III: The LLM (Gemini) generated CNF accurately represents a 2-bit multiplier without errors.")
                print("-" * 65)

    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found. Please check the file name and path.")
    except KeyError as e:
        print(f"Error: Column {e} not found in the CSV. Please ensure headers are exactly A0,A1,B0,B1,P0,P1,P2,P3.")

# 这里已经改成了你的文件名！
validate_truth_table('mul_2bit_vali_gem.csv')
```
